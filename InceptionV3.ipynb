```python
from keras.applications.inception_v3 import InceptionV3
import numpy as np
import os
import cv2
from PIL import Image
from keras.preprocessing.image import img_to_array
```

```python
path = "/home/shariva/Desktop/B.Tech/project/chota_test_dataset/102455176_5f8ead62d5.jpg"
image = Image.open(path)
print(image.size)
image = image.resize((299,299))
display(image)
image = img_to_array(image)
```

```python
model = InceptionV3(weights='imagenet', include_top=False)
features = model.predict(np.array([image]))
print(np.shape(features))
features
```

```python
directory = "/home/shariva/Desktop/B.Tech/project/chota_test_dataset/"
all_features = []
for image in os.listdir(directory):
    image = Image.open(os.path.join(directory,image))
    image = image.resize((299,299))
    display(image)
    image = img_to_array(image) #converts image to array of floats

    features = model.predict(np.array([image]))
    print(np.shape(features))
    print(features)
    all_features.append(features)
```
