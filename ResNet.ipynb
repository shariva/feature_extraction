```python
from keras.applications.resnet50 import ResNet50
from keras.preprocessing.image import img_to_array
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import os
```

```python
path = "/home/shariva/Desktop/B.Tech/project/chota_test_dataset/12830823_87d2654e31.jpg"
image = Image.open(path)
print(image.size)
image = image.resize((224,224))
display(image)
image = img_to_array(image) #converts image to array of floats
```

```python
model = ResNet50(weights='imagenet', include_top=False)
features = model.predict(np.array([image]))
print(np.shape(features))
features
```

```python
directory = "/home/shariva/Desktop/B.Tech/project/chota_test_dataset/"
all_features = []
for image in os.listdir(directory):
    image = Image.open(os.path.join(directory,image))
    image = image.resize((224,224))
    display(image)
    image = img_to_array(image) #converts image to array of floats

    features = model.predict(np.array([image]))
    print(np.shape(features))
    print(features)
    all_features.append(features)
```
